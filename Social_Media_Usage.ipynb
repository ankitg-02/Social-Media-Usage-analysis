```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing data from csv file
d=pd.read_csv(r'/content/Time-Wasters on Social Media.csv')
data=pd.DataFrame(d)
data.head()
```

**EDA based on time factor**

```python
plt.figure(figsize=(10,10))
sns.displot(data['Time Spent On Video'])
plt.title('Distribution of Time Spent on Video')
plt.show()
```

```python
plt.figure(figsize=(13,7))
sns.lineplot(x=data['Watch Time'],y=data['Time Spent On Video'])
plt.title('Watch Time vs Time Spent on Video')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x=data['Profession'],y=data['Total Time Spent'])
plt.title('Total Time Spent by Professionals')
plt.show
```

```python
sns.lineplot(x=data['Time Spent On Video'],y=data['UserID'],color='red',label="count of user")
plt.title('Time spend on video by users')
plt.show()
```

```python
sns.lineplot(x=data['Number of Videos Watched'],y=data['UserID'],label="count of user",color="green")
plt.title('Number of Videos Watched')
plt.show()
```

```python
sns.lineplot(x=data['Video Length'],y=data['UserID'],label="count of user")
plt.title('Video Duration')
plt.show()
```
